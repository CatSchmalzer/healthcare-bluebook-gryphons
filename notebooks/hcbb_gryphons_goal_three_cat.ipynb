## practice reading in 1000 rows

```python
%%time

procedures_provider = pd.DataFrame()
chunk_nbr = 1

for chunk in pd.read_csv('../data/Medicare_Provider_Utilization_and_Payment_Data__Physician_and_Other_Supplier_PUF_CY2017.csv',
                         dtype={'Zip Code of the Provider': object},
                         chunksize = 1000):
    procedures_provider.append(chunk[chunk['Entity Type of the Provider'] == 'I']) 
    
    if chunk_nbr == 1:
        procedures_provider = procedures_provider.append(chunk)
        chunk_nbr += 1
    else: break

procedures_provider.head()
```

## Read in datafile with 'Entity Type of the Provider' = 'I'

```python
%%time

HCPCS_providers_base =[]
for chunk in pd.read_csv('../data/Medicare_Provider_Utilization_and_Payment_Data__Physician_and_Other_Supplier_PUF_CY2017.csv', 
                         dtype={'Zip Code of the Provider': object},
                         chunksize = 1000):
    HCPCS_providers_base.append(chunk[chunk['Entity Type of the Provider'] == 'I']) 
               
                
hcpcs_providers = pd.concat(HCPCS_providers_base, ignore_index=True)
```

```python
hcpcs_providers.shape
```

```python
hcpcs_providers.head()
```

```python
hcpcs_providers.info()
```

```python
hcpcs_providers_clean = hcpcs_providers.drop(columns = ['Middle Initial of the Provider', 'Gender of the Provider', 'Street Address 2 of the Provider'])
```
